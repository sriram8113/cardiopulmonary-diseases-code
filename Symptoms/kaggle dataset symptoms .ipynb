```python
#removing diseases column to tranform the dataset

symptoms_data  = df.iloc[:,1:18]
symptoms_data.head()    
```

```python
#storing all the symptoms into list for further transformation

symptoms= []
for i in symptoms_data.columns:
    symptoms.append(symptoms_data[i].unique())

symptoms
    
```

```python
#storing all the unique symptoms together

symptoms_unq=()
for item in symptoms:
    symptoms_unq = list(set(symptoms_unq) | set(item))
print(symptoms_unq)
```

```python
#Removing nan values from unique symptoms list 

symptoms_unq = [x for x in symptoms_unq if x != 'nan']
```

```python
#checking no of unique symptoms

len(symptoms_unq)
```

```python
#dropping duplicate values 

df.drop_duplicates(inplace = True)
```

```python
#creating a new column named symptoms for transformation 

df['symptoms'] = 0
```

```python
df.head()
```

```python
#converting the new created column into string

df['symptoms'] = df['symptoms'].apply(lambda x: str(x))
```

```python
#chceking whether the new added column is converted to object or not

df.info()
```

```python
#storing all the columns data into symptoms column 

for i in range(df.shape[0]):
    values = df.iloc[i].values.tolist()
    if 0 in values:
        df["symptoms"][i] = values[1:values.index(0)]
    else:
        df["symptoms"][i] = values[1:]
    
    
    
```

```python
df.head()
```

```python
#creating a list having all the values of symtpoms

column_values = df[['Symptom_1', 'Symptom_2', 'Symptom_3', 'Symptom_4',
       'Symptom_5', 'Symptom_6', 'Symptom_7', 'Symptom_8', 'Symptom_9',
       'Symptom_10', 'Symptom_11', 'Symptom_12', 'Symptom_13', 'Symptom_14',
       'Symptom_15', 'Symptom_16', 'Symptom_17']].values.ravel()

```

```python
#storing unique values of symptoms into symps

symps = pd.unique(column_values)
symps = symps.tolist()
symps = [i for i in symps if str(i) != "nan"]
```

```python
symps
```

```python
#creating a datafram with unique values of symptoms as columns

symptoms = pd.DataFrame(columns = symps,index = df.index)
symptoms
```

```python
#Storing all the symptoms of a particukar disease into one column  and replacing nan values with 1 and 0 for specific diesese symptoms

symptoms["Symptoms"] = df["symptoms"]
for i in symps:
    symptoms[i] = symptoms.apply(lambda x:1 if i in x.Symptoms else 0, axis=1)
```

```python
symptoms
```

```python
#creating disease column and removing symptoms column 

symptoms["Disease"] = df["Disease"]
symptoms = symptoms.drop("Symptoms",axis=1)
symptoms.head()
```

```python
#storing the transformed dataset into new csv file 

symptoms.to_csv('kaggle_symptoms_cleaned.csv', index = False)
```
